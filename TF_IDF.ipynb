```python
import numpy as np
import pandas as pd
import csv
```

```python
df = pd.read_csv("wiki_movie_plots.csv", on_bad_lines='skip')
```

```python
title = df.iloc[:, 1].values
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='')
df.iloc[:, [4]] = imputer.fit_transform(df.iloc[:, [4]])
```

```python
df.columns = [c.replace(' ', '_') for c in df.columns]
```

```python
sentences = df['Release_Year'].apply(str) + ' ' + df['Title'] + ' ' + df['Origin/Ethnicity'] + ' ' + df['Director'] + ' ' + df['Cast'] + ' ' + df['Genre'] + ' ' + df['Plot']
```
